3- **Boxplot: Annual Premium by Vehicle Age**

The plot_premium_by_vehicle_age function returns a table that highlights essential information for the 'Annual_Premium' variable for each 'Vehicle_Age' category. Each column in the table is explained below:

1- Vehicle_Age: This column displays the many categories of the 'Vehicle_Age' variable.

4- **Correlation Heatmap: Pairwise Correlations**

The dataset yielded two pairs of strongly associated characteristics. Each pair's correlation coefficient is as follows:

The difference between Annual_Premium_normalized and Annual_Premium is 1.0000000000000016.

This shows a perfect positive correlation, indicating that these two characteristics are linearly related. One of them might be dropped without losing any information.
-0.8230971293403598 between Vehicle_Damage_encoded and Previously_Insured

This shows a significant inverse relationship. A negative correlation indicates that when one variable rises, the other tends to fall. The correlation in this situation implies an inverse link between Vehicle_Damage_encoded and Previously_Insured.
It's worth noting that perfect correlation (correlation coefficient of 1 or -1) might suggest that two properties are identical or that one is a linear transformation of the other.


```python
def plot_pairwise_correlations(data, correlation_threshold=0.8):
    """
    Plot a heatmap of pairwise correlations and print highly correlated feature pairs.

    Parameters:
    - data: DataFrame
        The input DataFrame.
    - correlation_threshold: float, optional
        The correlation coefficient threshold to identify highly correlated features.

    Returns:
    - None
    """
    # Calculate the correlation matrix
    correlation_matrix = data.corr()

    # Plot heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('Pairwise Correlations')
    plt.show()

    # Find highly correlated features
    highly_correlated_features = set()

    for i in range(len(correlation_matrix.columns)):
        for j in range(i):
            if abs(correlation_matrix.iloc[i, j]) > correlation_threshold:
                colname = correlation_matrix.columns[i]
                colname_related = correlation_matrix.columns[j]
                highly_correlated_features.add((colname, colname_related, correlation_matrix.iloc[i, j]))

    # Sort by absolute correlation coefficient
    highly_correlated_features = sorted(highly_correlated_features, key=lambda x: abs(x[2]), reverse=True)

    # Print the top 3 highly correlated features
    for feature_pair in highly_correlated_features[:3]:
        print(f"{feature_pair[0]} and {feature_pair[1]}: {feature_pair[2]}")

# Example usage
plot_pairwise_correlations(data)
```

Govt_job: There were no strokes among the 624 people who worked for the government. Stroke: 33 people working for the government had a stroke.

Never_worked: No Stroke: 22 people who had never worked did not get a stroke. Stroke: 0 people who have never worked have had a stroke.

**Insights:**

The table below breaks down stroke incidence by work type. It aids in comprehending the distribution of strokes across various labor types. Individuals with private occupations, for example, had a greater stroke rate than those with government positions. The dataset contains no strokes from the "Never_worked" category.
This data may be used to detect potential links between work kinds and the occurrence of strokes in the dataset.




```python
def plot_stroke_occurrences_by_work_type(data):
    # Count plot for stroke occurrences by work type
    plt.figure(figsize=(10, 6))
    sns.countplot(x='work_type', hue='stroke', data=data, palette='pastel')
    plt.title('Stroke Occurrences by Work Type')
    plt.xlabel('Work Type')
    plt.ylabel('Count')
    plt.show()

    # Display table of values
    stroke_by_work_type_table = pd.crosstab(data['work_type'], data['stroke'])
    stroke_by_work_type_table.columns = ['No Stroke', 'Stroke']
    stroke_by_work_type_table.reset_index(inplace=True)
    print(stroke_by_work_type_table)


plot_stroke_occurrences_by_work_type(data)
```
