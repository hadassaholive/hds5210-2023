### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
df = df.dropna(axis=1, how="all")
df = df.dropna(axis=0)
df.shape
```

```python
dis = []
for column in df:

    values = df[column].unique()
    count = len(values)
    dis.append(count)

dis.sort()
dis[-3: ]
```

```python
dic = {}
for column in df.columns:
    dic[column] = df[column].nunique()
reversed_dict = {value: key for key, value in dic.items()}

dict(sorted(reversed_dict.items()))
```

### 3. Data Redundancy

Use Python to identify any columns in your data where the value for every row is the same.  For example, if you had a gender column and all the rows said 'M', then you would want to write the code that helped you determine that.

```python
for column, unique_values in df.iteritems():
    if unique_values.nunique() == 1:
        print(f"{column} containing values that are same: {unique_values.iloc[0]}")
    else:
        print("Not every row has the same value")
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing columns like 'Country', 'Area', 'Density'

# Plotting histogram for 'Country'
plt.figure(figsize=(10, 5))
plt.hist(df['Country/Territory'], bins=30, edgecolor='black')
plt.title('Growth in different countries')
plt.xlabel('Country/Territory')
plt.ylabel('growth')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability if needed
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

selected_columns = ['Rank', 'CCA3', 'Country/Territory', 'Continent', '2022 Population', 'Area (km²)', 'Density (per km²)', 'Growth Rate', 'World Population Percentage']
print(df[selected_columns].describe())

plt.figure(figsize=(8, 6))
sns.scatterplot(x='2022 Population', y='Density (per km²)', data=df)
plt.title('2022 Population vs Density')
plt.xlabel('2022 Population')
plt.ylabel('Density (per km²)')
plt.show()

top_10_density = df.nlargest(10, 'Density (per km²)')
plt.figure(figsize=(10, 6))
sns.barplot(x='Country/Territory', y='Density (per km²)', data=top_10_density)
plt.title('Top 10 Countries by Population Density')
plt.xlabel('Country/Territory')
plt.ylabel('Density (per km²)')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Continent', y='Growth Rate', data=df)
plt.title('Population Growth Rate by Continent')
plt.xlabel('Continent')
plt.ylabel('Growth Rate')
plt.xticks(rotation=45)
plt.show()
```
